```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('data_ pre_processing/result_semester1_2016.xlsx')
```

```python
df
```

```python
# Group the data by alpha and calculate the mean for recall, precision, and F1-Score
grouped_data = df.groupby('alpha').mean()

# Extract the alpha values and the corresponding scores
alpha = grouped_data.index
recall = grouped_data['recall']
precision = grouped_data['precision']
f1_score = grouped_data['F1-Score']

# Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot recall, precision, and F1-Score
ax.plot(alpha, recall, marker='o', label='Recall')
ax.plot(alpha, precision, marker='o', label='Precision')
ax.plot(alpha, f1_score, marker='o', label='F1-Score')

# Set plot title and labels
ax.set_title('Average Scores for Different alpha Values')
ax.set_xlabel('Alpha')
ax.set_ylabel('Score')

# Set the legend
ax.legend()

# Display the plot
plt.show()
```

```python
# Group the data by beta and calculate the mean for recall, precision, and F1-Score
grouped_data = df.groupby('beta').mean()

# Extract the alpha values and the corresponding scores
alpha = grouped_data.index
recall = grouped_data['recall']
precision = grouped_data['precision']
f1_score = grouped_data['F1-Score']

# Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot recall, precision, and F1-Score
ax.plot(alpha, recall, marker='o', label='Recall')
ax.plot(alpha, precision, marker='o', label='Precision')
ax.plot(alpha, f1_score, marker='o', label='F1-Score')

# Set plot title and labels
ax.set_title('Average Scores for Different beta Values')
ax.set_xlabel('Beta')
ax.set_ylabel('Score')

# Set the legend
ax.legend()

# Display the plot
plt.show()
```

```python
# Group the data by gama and calculate the mean for recall, precision, and F1-Score
grouped_data = df.groupby('gama').mean()

# Extract the beta values and the corresponding scores
alpha = grouped_data.index
recall = grouped_data['recall']
precision = grouped_data['precision']
f1_score = grouped_data['F1-Score']

# Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot recall, precision, and F1-Score
ax.plot(alpha, recall, marker='o', label='Recall')
ax.plot(alpha, precision, marker='o', label='Precision')
ax.plot(alpha, f1_score, marker='o', label='F1-Score')

# Set plot title and labels
ax.set_title('Average Scores for Different Gama Values')
ax.set_xlabel('Gama')
ax.set_ylabel('Score')

# Set the legend
ax.legend()

# Display the plot
plt.show()
```

=> Hệ số alpha và beta có vẻ quan trong hơn so với hệ số gama. Điều đó có nghĩa là tiêu chí dựa theo sở thính và dựa theo điểm số tỉ lệ thuận với giá trị của recall, precision và F1-score, trong khi đó tiêu chí lựa chọn môn theo kỳ học có vẻ tỉ lệ nghịch

```python
# Sort the DataFrame by recall in descending order and select the top 10 rows
top_10_recall = df.nlargest(10, 'recall')

# Extract the alpha, beta, gamma, and recall values
alpha_values = top_10_recall['alpha']
beta_values = top_10_recall['beta']
gamma_values = top_10_recall['gama']
recall_values = top_10_recall['recall']

# Create a bar plot for the top 10 alpha, beta, gamma values with highest recall
plt.figure(figsize=(10, 6))
plt.bar(range(10), recall_values)
plt.xticks(range(10), [f'Alpha: {alpha}\nBeta: {beta}\nGamma: {gamma}' for alpha, beta, gamma in zip(alpha_values, beta_values, gamma_values)], rotation=45)
plt.xlabel('Alpha, Beta, Gamma')
plt.ylabel('Recall')
plt.title('Top 10 Alpha, Beta, Gamma with Highest Recall')
plt.tight_layout()


# Display the plot
plt.show()

```

==> Có thể suy ra được từ biểu đồ, hệ số beta có ảnh hướng rất lớn đến độ đo recall, khi giá trị beta có giá trị cao thì kết quả recall cũng ra kết quả càng cao
