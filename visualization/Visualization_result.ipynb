```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('data_ pre_processing/result_semester1_2016.xlsx')
```

```python
df
```

```python
# Group the data by alpha and calculate the mean for recall, precision, and F1-Score
grouped_data = df.groupby('alpha').mean()

# Extract the alpha values and the corresponding scores
alpha = grouped_data.index
recall = grouped_data['recall']
precision = grouped_data['precision']
f1_score = grouped_data['F1-Score']

# Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot recall, precision, and F1-Score
ax.plot(alpha, recall, marker='o', label='Recall')
ax.plot(alpha, precision, marker='o', label='Precision')
ax.plot(alpha, f1_score, marker='o', label='F1-Score')

# Set plot title and labels
ax.set_title('Average Scores for Different alpha Values')
ax.set_xlabel('Alpha')
ax.set_ylabel('Score')

# Set the legend
ax.legend()

# Display the plot
plt.show()
```

```python
# Group the data by beta and calculate the mean for recall, precision, and F1-Score
grouped_data = df.groupby('beta').mean()

# Extract the alpha values and the corresponding scores
alpha = grouped_data.index
recall = grouped_data['recall']
precision = grouped_data['precision']
f1_score = grouped_data['F1-Score']

# Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot recall, precision, and F1-Score
ax.plot(alpha, recall, marker='o', label='Recall')
ax.plot(alpha, precision, marker='o', label='Precision')
ax.plot(alpha, f1_score, marker='o', label='F1-Score')

# Set plot title and labels
ax.set_title('Average Scores for Different beta Values')
ax.set_xlabel('Beta')
ax.set_ylabel('Score')

# Set the legend
ax.legend()

# Display the plot
plt.show()
```
